```python
import pandas as pd
import numpy as np
import os
```

```python
def split_date_year(my_str):
    return int(my_str.split(" / ")[0])
    
def split_date_month(my_str):
    return int(my_str.split(" / ")[1])
```

```python
path = "DSNY_Monthly_Tonnage_Data.csv"
tonnage_csv = pd.read_csv(path)
tonnage_df = pd.DataFrame(tonnage_csv)
tonnage_df.head()
```

```python
tonnage_df = tonnage_df[["MONTH", "BOROUGH", "COMMUNITYDISTRICT", "REFUSETONSCOLLECTED", "PAPERTONSCOLLECTED", "MGPTONSCOLLECTED"]]
tonnage_df = tonnage_df.rename(columns={"MONTH": "Year/Month", 
                           "BOROUGH": "Borough",
                           "COMMUNITYDISTRICT": "CD",
                           "REFUSETONSCOLLECTED": "Refuse", 
                           "PAPERTONSCOLLECTED": "Paper",
                           "MGPTONSCOLLECTED": "MGP"})

tonnage_df["Year"] = tonnage_df['Year/Month'].map(split_date_year)
tonnage_df["Month"] = tonnage_df["Year/Month"].map(split_date_month)

tonnage_df = tonnage_df[["Year/Month", "Year", "Month", "Borough", "CD", "Refuse", "Paper", "MGP"]]
```

```python
tonnage_gb = tonnage_df.groupby("Year")
tonnage_gb.count().head()
```

```python
tonnage_df_clean = tonnage_df[tonnage_df["Year"] > 1991]

tonnage_df_clean["Month Index"] = (tonnage_df_clean["Year"] - 1992) * 12 + tonnage_df_clean["Month"]

tonnage_df_clean = tonnage_df_clean[["Month Index", "Year/Month", "Year", "Month", "Borough", "CD", "Refuse", "Paper", "MGP"]]
```
